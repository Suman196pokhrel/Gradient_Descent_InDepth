```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

## Lets use OLS so that we would know the value of m, b prior to our implementation of gradient descent

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_, reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='orange')
```

## Now we are going to write code , starting with a random value with the gradient descent to find the optimal value at the end
### Our end goal would be to start from any random value and reach 26.15 for our intercept value

```python
# lets apply gradient descent assuming that slope is constant m = 78.35
# and starting value for our intercept b = 0

y_pred = ((78.35*x) + 0).reshape(4)
y_pred
```

#### For now we have y values for all datapoint , enought to plot a line, whcih is obvious to be incorrect , but lets observe

```python
plt.scatter(x,y)
# Optimal LIne
plt.plot(x, reg.predict(x),color='red',label='OLS')

#Line that we have right now from b =0
plt.plot(x, y_pred,color='green',label="b = 0")
plt.legend()
```

### Now we have to apply gradient descent and see how the algorithm finds optimal value for b

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
# assuming learning rate to be
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# now calculating new intercept
b = b - step_size
b
```

```python
# Again calculating new y_pred and plotting new line
y_pred1 = ((78.35*x) + b).reshape(4)

plt.scatter(x,y)
# Optimal LIne
plt.plot(x, reg.predict(x),color='black',label='OLS')
plt.plot(x, y_pred1,color='orange',label=f'b = {b}')
plt.plot(x, y_pred,color='green',label=f'b = 0')

plt.legend()
```

## Now writing the above concepts in a loop

```python
b = -100
m = 78.35
lr = 0.01

epochs = 30
plt.figure(figsize=(15,15))

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*x.ravel() -b)
    b = b - (lr * loss_slope)
    y_pred = m * x + b
    
    plt.plot(x, y_pred)

plt.scatter(x,y)
```
