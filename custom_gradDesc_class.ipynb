```python
%matplotlib inline
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = make_regression(n_samples=300, n_features=1,n_informative=1,n_targets=1,noise=70,random_state=43)
```

```python
plt.figure(figsize=(9,7))
t = np.arange(len(x))
plt.scatter(x,y,c=t,s=50)
```

### Lets first get the actuial results

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_, reg.intercept_
```

```python
plt.figure(figsize=(9,7))
t = np.arange(len(x))
plt.scatter(x,y,c=t,s=50,alpha=0.4)

# Plotting the actual line for best fit given by sklearn's model
plt.plot(x, reg.predict(x),color="red")
```

```python
class GDRefressor:
    def __init__(self, learning_rate,epochs):
        self.m = 77.68
        self.b = -120 # initial arbitary value
        self.lr = learning_rate
        self.epochs = epochs
    
    def fit(self,x,y):
        # calculate the b using gradient descent
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m * x.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
#             print(f"Loss_Slope : {loss_slope}, B : {self.b}")
        print(self.b)
    
```

```python
gd = GDRefressor(learning_rate=0.001, epochs=100)
gd.fit(x,y)
```
