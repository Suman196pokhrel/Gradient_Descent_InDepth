```python
x, y = make_regression(n_samples=300, n_features=1,n_informative=1,n_targets=1,noise=70,random_state=43)
```

```python
plt.figure(figsize=(9,7))
t = np.arange(len(x))
plt.scatter(x,y,c=t,s=50)
```

```python
class GDRefressor:
    def __init__(self, learning_rate,epochs):
        self.m = 77.68
        self.b = -220 # initial arbitary value
        self.lr = learning_rate
        self.epochs = epochs
    
    def fit(self,x,y):
        # calculate the b using gradient descent
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m * x.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
            
            

        print(self.b)
    
    def predict(self,x):
        return self.m*x+ self.b
    
    def visualize(self,x,y,learning_rate,epochs,m,b):
        self.b = b
        self.m = m
        self.learning_rate = learning_rate
        self.epochs = epochs

       
        plt.figure(figsize=(11,9))
        
        t = np.arange(len(x))
        plt.scatter(x,y,c=t,s=50,alpha=0.3)
        

        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - m*x.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
            y_pred = self.m * x + self.b

            plt.plot(x, y_pred)
        print(self.b)
        
        
    
```

```python
gd = GDRefressor(learning_rate=0.001, epochs=100)
gd.fit(x,y)

```

```python
gd.predict(x)
```

```python
gd.visualize(x,y,learning_rate=0.001, epochs=100,m=77.68,b =-120)
```
